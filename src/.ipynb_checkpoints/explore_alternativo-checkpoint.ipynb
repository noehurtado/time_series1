```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('../data/raw/sales.csv')
df.head(3)
```

```python
df['date'] = pd.to_datetime(df['date'])
df = df.set_index('date')
df.head(3)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**Questions**
- Which is the tensor of the time serie?
- Which is the trend?
- Is it stationary?
- Is there variability or noise?

#### Trend analysis

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomp = seasonal_decompose(df, period = 12)

plt.figure(figsize=(10,5))
sns.lineplot(df)
sns.lineplot(decomp.trend)
plt.tight_layout()
plt.show()
```

#### Seasonality analysis

```python
plt.figure(figsize=(10,5))
sns.lineplot(df)
sns.lineplot(decomp.seasonal)
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

def dickey_fuller_test(df):
    print('Dickey-Fuller test results:')
    dftest = adfuller(df, autolag = "AIC")
    adf_test = pd.Series(dftest[0:4], index=["Test Statistic:", "p-value:", "#Lags Used:", "Number of Observations Used:"])
    for key, value in dftest[4].items():
        adf_test[f'Critical value {key}:'] = value
    return adf_test
```

```python
dickey_fuller_test(df)
```

#### Variability Analysis

```python
residual = decomp.resid

plt.figure(figsize=(10,5))
sns.lineplot(df)
sns.lineplot(decomp.resid)
plt.tight_layout()
plt.show()
```

#### Autocorrelation analysis

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(df)
plt.tight_layout()
plt.show()
```
